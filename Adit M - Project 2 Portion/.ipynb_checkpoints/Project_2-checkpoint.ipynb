```python
def checkcombinedcsv():
    inputPath = os.path.join('Data.csv')
    with open(inputPath) as f:
        next(f)
        Top_Property = []
        Top_City = []
        Top_Price = []
        Top_Location = []
        for line in f:
            line = line.strip()
            splitElements = line.split(',')
            PropertyType = splitElements[2]
            Address = splitElements[3]
            City = splitElements[4]
            ZipCode = splitElements[6]
            Price = splitElements[7]
            Location = [splitElements[-2], splitElements[-1]]
            Top_Property.append(PropertyType)
            Top_City.append(City)
            Top_Price.append(Price)
            Top_Location.append(Location)
            Real_Estate_Dict = {
                "Location": Location,
                "Property_Type": PropertyType,
                "City": City,
                "Zip_Code": ZipCode,
                "Price": Price
            }
            print(Real_Estate_Dict)
```

```python
checkcombinedcsv()
```
